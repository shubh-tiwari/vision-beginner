# Creating pencil sketch with OpenCV and Python

Using OpenCV, an image can be converted into a pencil sketch in four simple steps:

1. Convert the RGB color image to grayscale.
2. Invert the grayscale image to get a negative.
3. Apply a Gaussian blur to the negative from step 2.
4. Blend the grayscale image from step 1 with the blurred negative from step 3 using a color dodge.

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading image
img = 'sachin.jpg'
image = cv2.imread(img, 1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image.shape
```

```python
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
# Function to plot the image
def implot(img):
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
# Converting RGB to grayscale
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
implot(gray_img)
```

```python
# Obtaining the negative of the given image
inverted_img = 255-gray_img
implot(inverted_img)
```

```python
#  Applying Gaussian blur to negative image
blur_img = cv2.GaussianBlur(inverted_img, ksize=(25, 25),sigmaX=100, sigmaY=50)
implot(blur_img)
```

```python
# Blend the grayscale image with the blurred negative 
blend_img = cv2.divide(gray_img, 255-blur_img, scale=256)
implot(blend_img)
```

```python
cv2.imwrite('sachin_sketch.jpg', blend_img) 
```
